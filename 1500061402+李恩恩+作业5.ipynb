$$1、迎风法数值解一阶对流方程$$

$$a=-1,t:0\to4,x:-15\to1$$

```python
 #输出数据时用到了文件包“prettytable.py”
import numpy as np
import math
import matplotlib.pyplot as plt
from prettytable import PrettyTable
#定义迎风函数
def yf(a,dt,dx,t):#a<0,t为取出图像的时刻
    xfinal=-15
    xini=0
    tfinal=4
    tini=0
    nt=int(abs(tfinal-tini)/dt)+1
    nx=int(abs(xfinal-xini)/dx)+1
    r=a*dt/dx
    u1=[]
    for i in range(nx+2):
        for j in range(nt+1):
            u1.append(0.00)
    u=np.matrix(u1).reshape(nx+2,nt+1)
    #方波输入
    for i in range(nx+2):
        if(i>nx-6):
            u[i,0]=1
        else:
            u[i,0]=0
    for j in range(nt):
        for i in range(nx+1):
            u[i,j+1]=(1+r)*u[i,j]-r*u[i+1,j]
    m=int(t/dt)
    n=[]
    for i in range(nx+2):
        n.append(xfinal+i*dx)
    plt.ylim(0,1.5,0.01)
    plt.xlim(-15,1,0.5)
    plt.plot(n,u[:,m],c="r")
    return()
print("dt=0.04,dx=0.05")
for i in range(-1,4):
    yf(-1,0.04,0.05,i+1)#从右到左依次是时刻0s、1s、2s、3s、4s的图像 
```

```python
print("dt=dx=0.05")
for i in range(-1,4):
    yf(-1,0.05,0.05,i+1)#从右到左依次是时刻0s、1s、2s、3s、4s的图像
```

```python
print("dt=0.06,dx=0.05")
for i in range(-1,4):
    yf(-1,0.06,0.05,i+1)
```

```python
 #输出数据时用到了文件包“prettytable.py”
import numpy as np
import math
from scipy import stats
import matplotlib.pyplot as plt
from prettytable import PrettyTable
#定义迎风函数
def yf(a,dt,dx,t):#a<0,t为取出图像的时刻
    xfinal=-15
    xini=0
    tfinal=4
    tini=0
    nt=int(abs(tfinal-tini)/dt)+1
    nx=int(abs(xfinal-xini)/dx)+1
    r=a*dt/dx
    u1=[]
    for i in range(nx+2):
        for j in range(nt+1):
            u1.append(0.00)
    u=np.matrix(u1).reshape(nx+2,nt+1)
    #高斯波输入
    mu=0.05
    sigma=0.2
    x=[]
    for i in range(nx+2):
        x.append(xfinal+i*dx)
    for i in range(len(stats.norm.pdf(x,mu,sigma))):
        u[i,0]=stats.norm.pdf(x,mu,sigma)[i]
    for j in range(nt):
        for i in range(nx+1):
            u[i,j+1]=(1+r)*u[i,j]-r*u[i+1,j]
    m=int(t/dt)
    plt.ylim(0,2,0.01)
    plt.xlim(-15,1,0.5)
    plt.plot(x,u[:,m],c="r")
    return()
print("dt=0.04,dx=0.05")
for i in range(-1,4):
    yf(-1,0.04,0.05,i+1)#从右到左依次是时刻0s、1s、2s、3s、4s的图像
```

```python
print("dt=dx=0.05")
for i in range(-1,4):
    yf(-1,0.05,0.05,i+1)#从右到左依次是时刻0s、1s、2s、3s、4s的图像
```

```python
print("dt=0.06,dx=0.05")
for i in range(-1,4):
    yf(-1,0.06,0.05,i+1)
```

由前面所述得到：无论是高斯波还是方波输入，当稳定性条件$\Delta{t}<\Delta$x/|a|时，随着时间的增加，波形都变成高斯形状，并且波的高度降低、宽度增加，当稳定性条件$\Delta$t=$\Delta$x/|a|时，波形保持输入时的形状不变，当稳定性条件$\Delta$t>$\Delta$x/|a|时，方波或者是高斯波都会发散。
