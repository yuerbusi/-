## 1.数值误差的避免
### $\quad(a)$
给出计算$\bar{x}$，用样本数目$N$以及机器舍入误差精度$\dfrac{\epsilon_M}{2}$来表达.<br>
用浮点数表示样本的平均值为：
$$fl(\bar{x})=\left((\dots((x_1+x_2)(1+\varepsilon_1)+x_3)(1+\varepsilon_2)+
                     \dots+x_N)(1+\varepsilon_{N-1})\cdot\dfrac{1}{N}\right)(1+\varepsilon_N)$$
舍去高阶小项($\geq2阶$)，得：
\begin{equation*}
    fl(\bar{x})\approx \bar{x}\left(1+\dfrac{x_1+x_2}{\sum_{i=1}^{N}x_i}\varepsilon_1+
                                   \dfrac{x_1 + x_2 + x_3}{\sum_{i = 1}^N  x_i}\varepsilon_2+
                                   \dots+\varepsilon_{N-1} + \varepsilon_N \right)
    \end{equation*}
由于$|\varepsilon_i| \leq \dfrac{\epsilon_M}{2}$对$\forall 1\le i\le N$均成立，所以：
\begin{equation*}
    \begin{aligned}
    E(\bar{x})&=|\bar{x}-fl(\bar{x})|\\
    &\approx|\bar{x}|\cdot\left|\frac{x_1+x_2}{\sum_{i=1}^{N}x_i}\varepsilon_1+
    \frac{x_1+x_2+x_3}{\sum_{i=1}^N x_i}\varepsilon_2+\dots+\varepsilon_{N-1}+\varepsilon_N\right|\\
    &<|\bar{x}|\cdot|\varepsilon_1+\varepsilon_2+\dots+\varepsilon_N|\\
    &\le|\bar{x}|\cdot N\cdot\dfrac{\epsilon_M}{2}
    \end{aligned}
    \end{equation*}
故计算$\bar{x}$的舍入误差的最大可能的上限估计为:$$|\bar{x}|\cdot N\cdot \dfrac{\epsilon_M}{2}$$.

### $\quad(b)$
方差公式$[I]:\quad S^2=\dfrac{1}{N-1}\left(\sum\limits_{i=1}^N{x_i^2-N\cdot\bar{x}^2}\right)$，<br>
用浮点数表示的方差为：

\begin{equation*}
    \begin{aligned}
        fl(S^2)&=(((\dots((x_1^2(1+\varepsilon_1)+x_2^2(1+\varepsilon_2))\cdot(1+\varepsilon_3)\\
                         &+x_3^2(1+\varepsilon_4))\cdots+x_N^2(1-\varepsilon_{2N-2}))\cdot(1+\varepsilon_{2N-1})\\
                  &-N\cdot fl(\bar{x})^2\cdot(1+\varepsilon_{2N})\cdot(1+\varepsilon_{2N+1}))\\
                 &\cdot (1+\varepsilon_{2N+2})\cdot\dfrac{1}{N-1})(1+\varepsilon_{2N+3})
    \end{aligned}
    \end{equation*}
舍去高阶小项($\geq2阶$)，得：
\begin{equation*}
    \begin{aligned}
        fl(S^2)\approx&\dfrac{(\sum\limits_{i=1}^N x_i^2-N\cdot fl(\bar{x})^2)}{N-1}\cdot
                               (1+\dfrac{x_1^2}{\sum_{i=1}^N x_i^2-N\cdot fl(\bar{x}))^2}\cdot \varepsilon_1+\dfrac{x_2^2}{\sum_{i=1}^Nx_i^2-N\cdot fl(\bar{x}))^2}\cdot \varepsilon_2\\
                             &+\dfrac{x_1^2+x_2^2}{\sum_{i=1}^N x_i^2-N\cdot fl(\bar{x}))^2}\cdot \varepsilon_3+
                               \dots+\varepsilon_{2N+2}+\varepsilon_{2N+3})
    \end{aligned}
    \end{equation*}
由(a)得到：$fl(\bar{x})=\bar{x}(1+\varepsilon_0)$，且$|\varepsilon_0|<N\cdot\dfrac{\epsilon_M}{2}$，
    则$$fl(\bar{x})^2=\bar{x}^2(1+\varepsilon_0)^2\approx\bar{x}^2(1+2\varepsilon_0).$$
代入$fl(s^2)$，得：
    \begin{equation*}
    \begin{aligned}
    fl(S^2)\approx&(S^2-\dfrac{2N}{N-1}\varepsilon_0\bar{x}^2)(1+\frac{x_1^2}{\sum_{i=1}^N x_i^2-N\cdot fl(\bar{x})^2}\varepsilon_{1}+\frac{x_2^2}{\sum_{i=1}^{N}x_i^2-N\cdot fl(\bar{x})^2}\varepsilon_{2}\\
    &+\frac{x_1^2+x_2^2}{\sum_{i=1}^N x_i^2-N\cdot fl(\bar{x})^2}\varepsilon_3+\dots+\varepsilon_{2N+2}+\varepsilon_{2N+3})
    \end{aligned}
    \end{equation*}
    
又因为$|\varepsilon_i|\leq\dfrac{\epsilon_M}{2}$对$\forall 1\le i\le (2N+3)$均成立，所以误差为：
    \begin{equation*}
    \begin{aligned}
        E(S^2)=&|S^2-fl(S^2)|\\
        \approx&|S^2(\frac{x_1^2}{\sum_{i=1}^{N}x_i^2-N\cdot fl(\bar{x})^2}\varepsilon_{1}+
                     \frac{x_2^2}{\sum_{i=1}^{N}x_i^2-N\cdot fl(\bar{x})^2}\varepsilon_{2}
                     +\frac{x_1^2+x_2^2}{\sum_{i=1}^{N}x_i^2-N\cdot fl(\bar{x})^2}\varepsilon_{3}\\
        &+\dots+\varepsilon_{2N+2}+\varepsilon_{2N+3})-\dfrac{2N}{N-1}\varepsilon_{0}\bar{x}^2
                    (1+\frac{x_1^2}{\sum_{i=1}^{N}x_i^2-N\cdot fl(\bar{x})^2}\varepsilon_{1}\\
        &+\frac{x_2^2}{\sum_{i=1}^{N}x_i^2-N\cdot fl(\bar{x})^2}\varepsilon_{2}
                    +\frac{x_1^2+x_2^2}{\sum_{i=1}^{N}x_i^2-N\cdot fl(\bar{x})^2}\varepsilon_{3}
                    +\dots+\varepsilon_{2N+2}+\varepsilon_{2N+3})|\\
        \le&|S^2+\dfrac{2N}{N-1}\varepsilon_{0}\bar{x}^2|\cdot|\frac{x_1^2}{\sum_{i=1}^{N}x_i^2-N\cdot
                    fl(\bar{x})^2}\varepsilon_{1}+\frac{x_2^2}{\sum_{i=1}^{N}x_i^2-N\cdot fl(\bar{x})^2}\varepsilon_{2}\\
        &+\frac{x_1^2+x_2^2}{\sum_{i=1}^{N}x_i^2-N\cdot fl(\bar{x})^2}\varepsilon_{3}+\dots
                    +\varepsilon_{2N+2}+\varepsilon_{2N+3}|+|\dfrac{2N}{N-1}\varepsilon_{0}\bar{x}^2|\\
        \le&|S^2+\dfrac{2N}{N-1}\varepsilon_{0}\bar{x}^2|\cdot|\dfrac{\epsilon_M}{2}|\cdot
                    |\frac{x_1^2}{\sum_{i=1}^{N}x_i^2-N\cdot fl(\bar{x})^2}+\frac{x_2^2}{\sum_{i=1}^{N}x_i^2-N\cdot fl(\bar{x})^2}\\
        &+\frac{x_1^2+x_2^2}{\sum_{i=1}^{N}x_i^2-N\cdot fl(\bar{x})^2}+\dots+1+1|
                    +|\dfrac{2N}{N-1}\varepsilon_{0}\bar{x}^2|
    \end{aligned}
    \end{equation*}
由于$|\dfrac{x_1^2}{\sum\limits_{i=1}^N x_i^2-N\cdot fl(\bar{x})^2}+\dfrac{x_2^2}{\sum\limits_{i=1}^N x_i^2-N\cdot 
    fl(\bar{x})^2}+\dfrac{x_1^2+x_2^2}{\sum\limits_{i=1}^{N}x_i^2-N\cdot fl(\bar{x})^2}+\dots+1+1|$
    中除了常数1之外的每一项都有可能大于1且无上界,故其值无上界.
    
因此由公式$S^2=\dfrac{1}{N-1}(\sum\limits_{i=1}^N x_i^2-N\cdot\bar{x}^2)$计算得到的误差项$E(S^2)$无上界.

方差公式$[II]:\quad S^2=\dfrac{1}{N-1}\sum\limits_{i=1}^N({x_i-\bar{x}})^2$，<br>
用浮点数表示的方差为：
\begin{equation*}
    \begin{aligned}
    fl(S^2)=&((\dots((x_1-fl(\bar{x}))^2(1+\varepsilon_{1})^2(1+\varepsilon_{2})
                 +(x_2-fl(\bar{x}))^2(1+\varepsilon_{3})^2(1+\varepsilon_{4}))(1+\varepsilon_{5})\\
            &+\dots+(x_N-fl(\bar{x}))^2(1+\varepsilon_{3N-3})^2(1+\varepsilon_{3N-2}))(1+\varepsilon_{3N-1})
              \cdot\dfrac{1}{N-1})(1+\varepsilon_{3N})
    \end{aligned}
    \end{equation*}
舍去高阶($\geq2$阶)小项，得：
    \begin{equation*}
    \begin{aligned}
        fl(S^2)\approx&\dfrac{\sum\limits_{i=1}^{N}(x_i-fl(\bar{x}))^2}{N-1}
        (1+\frac{(x_1-fl(\bar{x}))^2}{\sum_{i=1}^{N}(x_i-fl(\bar{x}))^2}(2\varepsilon_{1}+\varepsilon_{2})
             +\frac{(x_2-fl(\bar{x}))^2}{\sum_{i=1}^{N}(x_i-fl(\bar{x}))^2}(2\varepsilon_{3}+\varepsilon_{4})\\
        &+\dots+\varepsilon_{3N-1}+\varepsilon_{3N})
    \end{aligned}
    \end{equation*}
由(a)可知：$fl(\bar{x})=\bar{x}(1+\varepsilon_0)$，且$|\varepsilon_{0}|<N\cdot\dfrac{\epsilon_M}{2}$，由此可得：
    \begin{equation*}
    \begin{aligned}
        fl(S^2)\approx&\dfrac{\sum\limits_{i=1}^{N}(x_i-\bar{x}(1+\varepsilon_{0}))^2}{N-1}
        (1+\frac{(x_1-fl(\bar{x}))^2}{\sum_{i=1}^{N}(x_i-fl(\bar{x}))^2}(2\varepsilon_{1}
            +\varepsilon_{2})+\frac{(x_2-fl(\bar{x}))^2}{\sum_{i=1}^{N}(x_i-fl(\bar{x}))^2}(2\varepsilon_{3}+\varepsilon_{4})\\
        &+\dots+\varepsilon_{3N-1}+\varepsilon_{3N})\\
        =&\dfrac{\sum\limits_{i=1}^{N}((x_i-\bar{x})^2-2(x_i-\bar{x})\bar{x}\varepsilon_{0}+
            \bar{x}^2\varepsilon_{0}^2)}{N-1}(1+\frac{(x_1-fl(\bar{x}))^2}{\sum_{i=1}^{N}(x_i-fl(\bar{x}))^2}(2\varepsilon_{1}+\varepsilon_{2})\\
        &+\frac{(x_2-fl(\bar{x}))^2}{\sum_{i=1}^{N}(x_i-fl(\bar{x}))^2}(2\varepsilon_{3}
            +\varepsilon_{4})+\dots+\varepsilon_{3N-1}+\varepsilon_{3N})\\
            \approx&\dfrac{\sum\limits_{i=1}^{N}((x_i-\bar{x})^2-2x_i\bar{x}\varepsilon_{0}
            +2\bar{x}^2\varepsilon_{0})}{N-1}(1+\frac{(x_1-fl(\bar{x}))^2}{\sum_{i=1}^{N}(x_i-fl(\bar{x}))^2}(2\varepsilon_{1}+\varepsilon_{2})\\
        &+\frac{(x_2-fl(\bar{x}))^2}{\sum_{i=1}^{N}(x_i-fl(\bar{x}))^2}(2\varepsilon_{3}+
            \varepsilon_{4})+\dots+\varepsilon_{3N-1}+\varepsilon_{3N})\\
        =&\dfrac{\sum\limits_{i=1}^{N}(x_i-\bar{x})^2}{N-1}(1+\frac{(x_1-fl(\bar{x}))^2}{\sum_{i=1}^{N}
            (x_i-fl(\bar{x}))^2}(2\varepsilon_{1}+\varepsilon_{2})+\frac{(x_2-fl(\bar{x}))^2}{\sum_{i=1}^{N}
            (x_i-fl(\bar{x}))^2}(2\varepsilon_{3}+\varepsilon_{4})\\
        &+\dots+\varepsilon_{3N-1}+\varepsilon_{3N})
    \end{aligned}
    \end{equation*}
又因为$|\varepsilon_i|\leq\dfrac{\epsilon_M}{2}$对$\forall 1\le i\le 3N$均成立，所以误差为：
    \begin{equation*}
    \begin{aligned}
        E(S^2)=&|S^2-fl(S^2)|\\
        \approx&S^2|\frac{(x_1-fl(\bar{x}))^2}{\sum_{i=1}^{N}(x_i-fl(\bar{x}))^2}(2\varepsilon_{1}
            +\varepsilon_{2})+\frac{(x_2-fl(\bar{x}))^2}{\sum_{i=1}^{N}(x_i-fl(\bar{x}))^2}(2\varepsilon_{3}
            +\varepsilon_{4})+\dots+\varepsilon_{3N-1}+\varepsilon_{3N}|\\
        <&S^2|(2\varepsilon_{1}+\varepsilon_{2})+(2\varepsilon_{3}+\varepsilon_{4})+
            \dots+\varepsilon_{3N-1}+\varepsilon_{3N}|\\
        \le&S^2\cdot 4N\cdot|\dfrac{\epsilon_M}{2}|
    \end{aligned}
    \end{equation*}
故由公式$S^2=\dfrac{1}{N-1}\sum\limits_{i=1}^{N}(x_i-\bar{x})^2$计算得到的误差项$E(S^2)$的上界为
    $S^2\cdot 4N\cdot|\dfrac{\epsilon_M}{2}|$.
        
综上所述，方差公式$[II]:\quad S^2=\dfrac{1}{N-1}\sum\limits_{i=1}^{N}(x_i-\bar{x})^2$
比方差公式$[I]:\quad S^2=\dfrac{1}{N-1}\left(\sum\limits_{i=1}^{N}x_i^2-N\cdot\bar{x}^2\right)$在数值计算中更为稳定和准确.

### $\quad(c)$说明矩阵$A$的无穷模形式
由题目可知矩阵$p$模的定义为：
    $$||A||_p = \sup\limits_{x\neq 0} \dfrac{||Ax||_p}{||x||_p}$$
当取$p\rightarrow\infty$时，矩阵的$\infty$模为：
    $$||A||_\infty = \sup\limits_{x\neq 0} \dfrac{||Ax||_\infty}{||x||_\infty}$$
$令向量b=A\cdot x，则其第i个分量b_i=\sum\limits_{j=1}^{n}a_{ij}x_{j}，
其中a_{ij}和x_{j}分别为矩阵A的(i,j)元和向量x的第j个分量.$
我们知道向量$p$模的定义为：
    $$||x||_p=\left(\sum\limits_{k=1}^{n}|x_k|^p\right)^{1/p}$$
则向量$b的\infty$模为：
    $$||b||_\infty=||Ax||_\infty=\max\limits_{i=1,2,\dots,n}\left|\sum\limits_{j=1}^{n}a_{ij}x_{j}\right|$$
令$x_{max}=\max\limits_{i=1,2,\dots,n}|x_k|>0$，可得：
    $$\dfrac{||Ax||_\infty}{||x||_\infty}
    =\dfrac{\max\limits_{i=1,2,\dots,n}\left|\sum\limits_{j=1}^{n}a_{ij}x_{j}\right|}{x_{max}}
    \leq \max\limits_{i=1,2,\dots,n}\sum\limits_{j=1}^{n}|a_{ij}|\dfrac{|x_j|}{x_{max}}
    \leq \max\limits_{i=1,2,\dots,n}\sum\limits_{j=1}^{n}|a_{ij}|.$$
    显然当取合适的向量$x$时，上式中的两个等号可以同时取到.

因此矩阵$A的\infty模为$:
    $$||A||_\infty = \sup\limits_{x\neq 0} \dfrac{||Ax||_\infty}{||x||_\infty}
    =\max\limits_{i=1,2,\dots,n}\sum\limits_{j=1}^{n}|a_{ij}|$$
### $\quad(d)$ 幺正矩阵的几个证明        
#### $\quad(d.1)$        
由于$U$为幺正矩阵，即$UU^\dagger=U^\dagger U=I$恒成立，则有：
\begin{equation*}

    ||U^\dagger x||_2&=\sqrt{(U^\dagger x,U^\dagger x)}=\sqrt{(U^\dagger x)^\dagger U^\dagger x}
=\sqrt{x^\dagger U U^\dagger x}=\sqrt{x^\dagger x}=\sqrt{(x,x)}=||x||_2
    \end{aligned}
    \end{equation*}
根据矩阵P模的定义$(p=2)$：
\begin{equation*}
    \begin{aligned}

    ||U^\dagger||_2&=\sup\limits_{x\neq 0}{\dfrac{||U^\dagger x||_2}{||x||_2}}=\sup\limits_{x\neq 0}{\dfrac{||x||_2}{||x||_2}}=1
    \end{aligned}
    \end{equation*}

## 3.Hilbert 矩阵
### (a)给出矩阵$H_n$的矩阵元的表达式和矢量$b$的表达式        
如果要让$\displaystyle{D=\int_0^1 dx\left(\sum_{j=1}^n c_j x^{j-1}-f(x)\right)^2}$取极小值，
则<font face="楷体" font size=3>必要条件</font>要求$D$对任意$x_i$的一阶偏导数均为零，即：
\begin{equation*}
    \begin{aligned}
    \dfrac{\partial D}{\partial c_i} = \int_0^1 2\left((\sum_{j=1}^n c_j x^{j-1} - f(x))x^{i-1}\mathrm d x \equiv 0\right)
    \end{aligned}
    \end{equation*}
即：
\begin{equation*}
    \begin{aligned}
    \sum_{j=1}^n c_j\int_0^1 x^{i+j-2} \mathrm d x&=\int_0^1 x^{i-1}f(x)\mathrm d x\\
    \sum_{j=1}^n \dfrac{1}{i+j-1}c_j&=\int_0^1 x^{i-1}f(x)\mathrm d x
    \end{aligned}
    \end{equation*}
将上式与各个系数$c_i$满足的方程:$\displaystyle{\sum_{j=1}^n [H_n]_{ij} c_j=b_i}$比较可得：
\begin{equation*}
    \begin{aligned}
    {[H_n]}_{ij}=\dfrac{1}{i+j-1}\\
    b_i=\int_0^1 x^{i-1}f(x)\mathrm dx
    \end{aligned}
    \end{equation*}
又因为：
\begin{equation*}
    \begin{aligned}
    \dfrac{\partial^2D}{\partial c_i\partial c_j}&=2\int_0^1x^{i+j-2}\mathrm dx=\dfrac{2}{i+j-1}\\
    &=2[H_n]_{ij}
    \end{aligned}
    \end{equation*}
且由$(b)$证得$Hilbert$矩阵为对称正定矩阵，因此$D$关于$c_i$的$Hessian$矩阵也应为对称正定矩阵，
那么，上述$c_i$满足的条件则上升为$D$取极小值的充分必要条件.        
### (b)证明矩阵$H_n$是对称正定矩阵，且为非奇异矩阵        
对于任意$c\in\mit R^n$，不妨设$c={(c_1,c_2,c_3,\dots,c_n)}$，
其第$i$个分量为$c_i$,其中$i=1,2,\dots,n.$
令$\mit Q_n$为一$(n-1)$次多项式，
定义为：$\displaystyle{\sum_{i=1}^n c_i x^{i-1}}$
则有：
\begin{gather*}
{\mit Q_n}(x)\equiv
\begin{pmatrix}
        c_1 & c_2 & c_3 & \dots & c_n\\
        \end{pmatrix}\cdot
\begin{pmatrix}
        1      \\
        x      \\
        x^2    \\
        \vdots \\
        x^{n-1}\\
        \end{pmatrix}\equiv
\begin{pmatrix}
        1 & x & x^2 & \dots & x^{n-1}\\
        \end{pmatrix}\cdot
\begin{pmatrix}
        c_1   \\
        c_2   \\
        c_3   \\
        \vdots\\
        c_n   \\
        \end{pmatrix}
\end{gather*}
那么：
\begin{gather*}
[{\mit Q_n}(x)]^2&=
\begin{pmatrix}
        c_1 & c_2 & c_3 & \dots & c_n\\
        \end{pmatrix}\cdot
\begin{pmatrix}
        1      \\
        x      \\
        x^2    \\
        \vdots \\
        x^{n-1}\\
        \end{pmatrix}\cdot
\begin{pmatrix}
        1 & x & x^2 & \dots & x^{n-1}\\
        \end{pmatrix}\cdot
\begin{pmatrix}
        c_1   \\
        c_2   \\
        c_3   \\
        \vdots\\
        c_n   \\
        \end{pmatrix}\\&=
\begin{pmatrix}
        c_1 & c_2 & c_3 & \dots & c_n\\
        \end{pmatrix}\cdot
\begin{pmatrix}
        1&     x  &      x^2&   \cdots&   x^{n-1} \\
        x&     x^2&      x^3&   \cdots&      x^n  \\
   \vdots&  \vdots&   \ddots&   \ddots&   \vdots  \\
  x^{n-1}&     x^n&  x^{n+1}&   \cdots&  x^{2n1-2}\\
        \end{pmatrix}\cdot
\begin{pmatrix}
        c_1   \\
        c_2   \\
        c_3   \\
        \vdots\\
        c_n   \\
        \end{pmatrix}
\end{gather*}
在区间$[0,1]$积分得：
\begin{gather*}
\int_0^1[{\mit Q_n}(x)]^2\mathrm dx=
\begin{pmatrix}
        c_1 & c_2 & c_3 & \dots & c_n\\
        \end{pmatrix}\cdot
\begin{pmatrix}
                   1&    \dfrac{1}{2}&    \dfrac{1}{3}&   \cdots&        \dfrac{1}{n}\\
        \dfrac{1}{2}&    \dfrac{1}{3}&    \dfrac{1}{4}&   \cdots&      \dfrac{1}{n+1}\\
              \vdots&          \vdots&          \ddots&   \ddots&              \vdots\\
       \dfrac{1}{n}&  \dfrac{1}{n+1}&  \dfrac{1}{n+2}&   \cdots&     \dfrac{1}{2n-1}\\
        \end{pmatrix}\cdot
\begin{pmatrix}
        c_1   \\
        c_2   \\
        c_3   \\
        \vdots\\
        c_n   \\
        \end{pmatrix}
\end{gather*}
即：
$$\int_0^1[{\mit Q_n}(x)]^2\mathrm dx=c^T\cdot H_n \cdot c\geq 0.$$
故$c^T\cdot H_n \cdot c\geq 0$对于任意$c\in\mit R^n$恒成立，
并且等号只有当$\mit Q_n \equiv 0$时会取得，即只有当$c\equiv 0$时会取得.

由线性代数知识，$H_n$为对称正定矩阵，那么其也必为非奇异矩阵.
### (c)估计出$\det(H_n)$，$n\leq 0$ 的数值       
对矩阵$H_n$的行列式表达式两边取对数，得：
$$\ln (\det(H_n))=4\ln(c_n)-\ln(c_{2n})$$
对$c_n$的表达式取对数,，得：
\begin{equation*}
    \begin{aligned}
    \ln(c_n)&=\ln(1!)+\ln(2!)+\dots+\ln((n-1)!)\\
    &=(n-1)\ln1+(n-2)\ln2+\dots+\ln(n-1)\\
    &=\sum_{i=1}^{n-1}(n-i)\ln i\\
    &\approx \int_1^n\mathrm dx(n-x)\ln x\\
    &=\left(nx(\ln x-1)-\dfrac{x^2}{4}(2\ln x-1)\right)\Bigg|_1^n\\
    &=\dfrac{n^2}{2}\ln n-\dfrac{3}{4}n^2+n-\dfrac{1}{4}
    \end{aligned}
    \end{equation*}
将$n$替换为$2n$，可得：
$$\ln(c_{2n})\approx 2n^2\ln n-2n^2\ln2-3n^2+2n-\dfrac{1}{4}.$$
代入$\ln (\det(H_n))=4\ln(c_n)-\ln(c_{2n})$，得：
$$\ln (\det(H_n))\approx-2n^2\ln2+2n-\dfrac{3}{4}$$
所以最终$\det(H_n)$可近似为：
$$\det(H_n)\approx\dfrac{\large{e}^{2n-\small{\dfrac{3}{4}}}}{4^{n^2}}.$$


```python
# 计算det(Hn)的估计值
import pandas as pd
import math

det = [math.exp(2*i-3/4)/4**(i**2) for i in range(1,11)]
data=[]
for n in range(1,11):
   data.append([n,float('%13.6e'% det[n-1])])

form = pd.DataFrame(data,columns = ['n', 'det(Hn)估计值'])
form
```

### (d)分别用$GEM$方法和$Cholesky$方法求解$Hilbert$矩阵表示的线性方程组

当$n=1,2,\dots,10$时分别用部分选主元的$GEM$方法和$Cholesky$方法求解线性方程组
$H_n\cdot x=b$，其中$b=(1,1,\dots,1)\in{\mit R}^n$，程序如下：

```python
for n in range(1,11):
    # 创建n维Hilbert矩阵与n维向量b
    h=[]
    b=[]
    for i in range(n):
        h.append([])
        b.append(1)
    for i in range(n):
        for j in range(n):
            h[i].append(1/(i+j+1))
    # 部分选主元的GEM方法
    for k in range(n-1):
        p = k
        h_max = h[k][k]
        # 在当前列中选主元，并确定其所在行序号
        for i in range(k,n):
            if abs(h[i][k])>abs(h_max):
                h_max=h[i][k]
                p=i
        # 交换当前行与最大主元所在的一行，对Hilbert矩阵与b向量都进行操作
        if p!=k:
            t=h[k]
            h[k]=h[p]
            h[p]=t
            t=b[k]
            b[k]=b[p]
            b[p]=t
        # 如果主元为零，则继续下一个循环
        if h[k][k]==0:
            continue
        # 计算当前列的乘数
        for i in range(k+1,n):
            h[i][k]=h[i][k]/h[k][k]
        # 对剩余的子矩阵进行变换，对Hilbert矩阵与b向量都进行操作
        for i in range(k+1,n):
            for j in range(k+1,n):
                h[i][j]=h[i][j]-h[i][k]*h[k][j]
            b[i]=b[i]-h[i][k]*b[k]
    # 使用回代法解出x
    for j in range(n-1,-1,-1):
        if h[j][j]==0:
            print('错误！矩阵奇异！')
            break
        b[j]=b[j]/h[j][j]
        for i in range(j):
            b[i]=b[i]-h[i][j]*b[j]
    # 打印输出结果
    print('当n=%d时，用GEM方法给出的解为：' % n)
    for i in range(n-1):
        print('%13.6e' % b[i], end='，')
    print('%13.6e' % b[n-1])
```

```python
for n in range(1,11):
    # 创建n维Hilbert矩阵与n维向量b
    h=[]
    b=[]
    for i in range(n):
        h.append([])
        b.append(1)
    for i in range(n):
        for j in range(n):
            h[i].append(1/(i+j+1))
    # Cholesky方法
    import math
    for k in range(n):
        h[k][k]=math.sqrt(h[k][k])
        for i in range(k+1,n):
            h[i][k]=h[i][k]/h[k][k]
        for j in range(k+1,n):
            for i in range(j,n):
                h[i][j]=h[i][j]-h[i][k]*h[j][k]
    # 使H矩阵成为对称矩阵，其下三角和上三角乘积即为Hilbert矩阵
    for i in range(n):
        for j in range(i + 1, n):
            h[i][j] = h[j][i]
    # 使用前代法和回代法解出x
    for j in range(n):
        if h[j][j]==0:
            print('错误！矩阵奇异！')
            break
        b[j]=b[j]/h[j][j]
        for i in range(j+1,n):
            b[i]=b[i]-h[i][j]*b[j]
    for j in range(n-1,-1,-1):
        if h[j][j]==0:
            print('错误！矩阵奇异！')
            break
        b[j]=b[j]/h[j][j]
        for i in range(j):
            b[i]=b[i]-h[i][j]*b[j]
    # 打印输出结果
    print('当n=%d时，用Cholesky方法给出的解为：'%n)
    for i in range(n-1):
        print('%13.6e'%b[i],end='，')
    print('%13.6e'%b[n-1])
```

可以看出分别用部分选主元的$GEM$方法和$Cholesky$方法求解$n$维$Hilbert$矩阵构成的线性方程组$H_n\cdot x=b$时：
    
当$n\le 8$时两种方法得到的解分量在保留7位有效数字时无差异；而在$n=9$时在两种方法的解分量在第7位有效数字有所差异；
当$n=10$时在两种方法的解分量在第5位有效数字有所差异，

故可见两种方法给出的解差异很小，几乎没有。
相比较而言，我认为$Cholesky$方法更为精确。因为其利用了$Hilbert$矩阵的对称正定性质，运算量减少了一半，从而也减少了运算过程中误差的累积.
