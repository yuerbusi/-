```python
#t=1/pi(sec)时刻图像
import numpy as np
import math
import matplotlib.pyplot as plt
from prettytable import PrettyTable
nt=150000
nx=150
xfinal=2.0
xini=0.0
tfinal=2.1
tini=0.0
dt=(tfinal-tini)/nt#时间步长
dx=(xfinal-xini)/nx#空间步长
u=np.zeros((nx+3,nt+1))#建立一个元素均为零的矩阵
for i in range(nx+1):
    u[i,0]=math.cos(math.acos(-1.0)*(xini+i*dx))
u[nx+1,0]=u[1,0]
u[nx+2,0]=u[2,0]
for j in range(1,nt+1):
    for i in range(2,nx+1):
        u[i,j]=u[i,j-1]-dt*(1.0/6.0/dx*(u[i+1,j-1]+u[i,j-1]+u[i-1,j-1])*(u[i+1,j-1]-u[i-1,j-1])+0.022**2*(u[i+2,j-1]-2.0*u[i+1,j-1]+2.0*u[i-1,j-1]-u[i-2,j-1])/2.0/dx**3)
    u[0,j]=u[0,j-1]-dt*(1.0/6/dx*(u[1,j-1]+u[0,j-1]+u[nx-1,j-1])*(u[1,j-1]-u[nx-1,j-1])+0.022**2*(u[2,j-1]-2.0*u[1,j-1]+2.0*u[nx-1,j-1]-u[nx-2,j-1])/2.0/dx**3)
    u[1,j]=u[1,j-1]-dt*(1.0/6/dx*(u[2,j-1]+u[1,j-1]+u[0,j-1])*(u[2,j-1]-u[0,j-1])+0.022**2*(u[3,j-1]-2.0*u[2,j-1]+2.0*u[0,j-1]-u[nx-2,j-1])/2.0/dx**3)
    u[nx+1,j]=u[1,j]
    u[nx+2,j]=u[2,j]
x=[xini+i*dx for i in range(nx+3)]
#光滑处理
v=np.copy(u[:,int(1/math.pi/dt)])
for i in range(1,nx+2):
    v[i]=(u[i-1,int(1/math.pi/dt)]+u[i,int(1/math.pi/dt)]+u[i+1,int(1/math.pi/dt)])/3
plt.plot(x,v,c="r")
```
