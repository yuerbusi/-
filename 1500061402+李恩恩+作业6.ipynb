```python
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
import random 
N=100000
step=1000
L=0.52
#
u=np.zeros((step,N))
#
E=[]
#
V=[]
for j in range(N):
    walk=0
    for i in range(step):
        a=random.uniform(0,1)
        walk+=-1 if a<L else 1
        u[i,j]=walk
for i in range(step):
    E.append(np.mean(u[i]))
    V.append(np.var(u[i]))
#
plt.figure(figsize=(15,15))
plt.xlabel("steps")
plt.ylabel("mean")
plt.plot(range(len(E)),E, color='red')
plt.show()
#
plt.figure(figsize=(15,15))
plt.xlabel("steps")
plt.ylabel("sigma")
plt.plot(range(len(V)),V, color='yellow')
plt.show()
#
sns.distplot(u[999])
```
