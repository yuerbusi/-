```python
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
import random 
N=100000
step=1000
L=0.52
#初始化矩阵
u=np.zeros((step,N))
#均值
E=[]
#方差
V=[]
for j in range(N):
    walk=0
    for i in range(step):
        a=random.uniform(0,1)
        walk+=-1 if a<L else 1
        u[i,j]=walk
for i in range(step):
    E.append(np.mean(u[i]))
    V.append(np.var(u[i]))
#绘均值图
plt.figure(figsize=(15,15))
plt.xlabel("steps")
plt.ylabel("mean")
plt.plot(range(len(E)),E, color='red')
plt.show()
#绘方差图
plt.figure(figsize=(15,15))
plt.xlabel("steps")
plt.ylabel("sigma")
plt.plot(range(len(V)),V, color='yellow')
plt.show()
#绘位置分布图
sns.distplot(u[999])
```

从上述结果可以看出L=0.52时，均值随着步数的增加而线性减少，最终在1000步时均值为1000*（0.48-0.25）=-40，方差随着步数的增加线性增加，最终1000步时方差为1000左右，第1000步时分布概率随位置的变化近似服从高斯分布，峰值在均值-40附近。
